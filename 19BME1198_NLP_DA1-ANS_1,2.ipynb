```python
from nltk.corpus import brown
brown.categories()

brown.words(categories='government')

brown.words(fileids=['ch15'])
brown.sents(categories=['news', 'editorial', 'reviews'])
news_text = brown.words(categories='government')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
     print(m + ':', fdist[m], end=' ')
```

```python
btwords = brown.tagged_words(categories='news')
cfd = nltk.ConditionalFreqDist(btwords)
conditions = cfd.conditions()

mono_tags = [condition for condition in conditions if len(cfd[condition]) == 1]
a= len(mono_tags)
b= len(conditions)
print(a)
proportion_mono_tags = a/b 

number_mono_tags = len(conditions) - len(mono_tags)


total_words = set(brown.words())
number_shared_words = [word for word in mono_tags if word in total_words]
print(total_words)
print(len(total_words))
```
